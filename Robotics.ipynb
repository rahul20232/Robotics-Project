```python
x1 = np.array(df_ds['x'].copy())
y1 = np.array(df_ds['y'].copy())
z1 = np.array(df_ds['z'].copy())

x2 = np.array(df_sod['x'].copy())
y2 = np.array(df_sod['y'].copy())
z2 = np.array(df_sod['z'].copy())

x3 = np.array(df_gd['x'].copy())
y3 = np.array(df_gd['y'].copy())
z3 = np.array(df_gd['z'].copy())

marker_size = 2
# Create a 3D scatter plot for the first set of data
trace1 = go.Scatter3d(x=x1, y=y1, z=z1, mode='markers', name='DSV SLAM',marker=dict(size=marker_size))

# Create a 3D scatter plot for the second set of data
trace2 = go.Scatter3d(x=x2, y=y2, z=z2, mode='markers', name='Before Loop Closures',marker=dict(size=marker_size))

trace3 = go.Scatter3d(x=x3, y=y3, z=z3, mode='markers', name='Ground Truth',marker=dict(size=marker_size))

# Create a layout with axis labels
layout = go.Layout(scene=dict(xaxis_title='X Label', yaxis_title='Y Label', zaxis_title='Z Label'))

# Create a figure with both traces
fig = go.Figure(data=[trace3,trace2, trace1], layout=layout)

# Show the interactive plot
fig.show()
```

```python

plt.plot(x3, z3, label='gnd')
# plt.plot(x2,z2, label='sod')
plt.plot(x1,z1, label='after')
plt.legend()
plt.show()
```

```python
indexes = list(df_ds['0'])
frames_x = []
frames_y = []
frames_z = []
for i in range(len(x3)):
    if(i in indexes):
        frames_x.append(x3[i])
        frames_y.append(y3[i])
        frames_z.append(z3[i])
```

```python
import numpy as np

def calculate_ate(x1, y1, z1, x2, y2, z2):
    # Ensure all lists have the same length
    assert len(x1) == len(y1) == len(z1) == len(x2) == len(y2) == len(z2), "All lists must have the same length"

    # Convert lists to NumPy arrays for easier calculations
    path1 = np.array([x1, z1]).T  # Transpose to get (N, 3) shape
    path2 = np.array([x2, z2]).T
    
#     offset_x = x1[0] - x2[0]
#     offset_y = y1[0] - y2[0]
#     offset_z = z1[0] - z2[0]
    offset_x = 0
    offset_y = 0
    offset_z = 0
    # Calculate the Euclidean distance between corresponding points
    errors = ((x1-x2 - offset_x)**2  +(y1-y2-offset_y)**2+ (z1-z2-offset_z)**2)**0.5

    # Calculate the mean absoluteerror
    ate = np.mean(errors)

    return ate

x = frames_x
y = frames_y
z = frames_z

x_ = x1
y_ = y1
z_ = z1

ate_value = calculate_ate(x, y, z, x_, y_, z_)
print(f"Absolute Trajectory Error: {ate_value}")
```
